```python
import librosa
import numpy as np
import librosa.display
```

```python
y, _ = librosa.load("./002270472.wav", sr=16000)
```

```python
y.shape
```

```python
y = y[::2]
```

```python
librosa.display.specshow(
    librosa.amplitude_to_db(np.abs(librosa.stft(y, n_fft=512)), ref=np.max),
    y_axis="linear",
    x_axis="time",
)
```

```python
y_8000 = librosa.resample(y, orig_sr=16000, target_sr=8000)
y_48000 = librosa.resample(y, orig_sr=16000, target_sr=48000)
```

```python
librosa.display.specshow(
    librosa.amplitude_to_db(np.abs(librosa.stft(y_48000, n_fft=512)), ref=np.max),
    y_axis="linear",
    x_axis="time",
)
```

```python
librosa.display.specshow(
    librosa.amplitude_to_db(np.abs(librosa.stft(y, n_fft=512)), ref=np.max),
    y_axis="linear",
    x_axis="time",
    sr=16000,
)
```

```python
librosa.display.specshow(
    librosa.amplitude_to_db(np.abs(librosa.stft(y_8000, n_fft=512)), ref=np.max),
    y_axis="linear",
    x_axis="time",
)
```
